```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('DATASETS/ResultadosElectorales_2023_Generales.csv')
```

```python
df.head()
```

```python
df=df.drop(['lista_numero', 'lista_nombre', 'recuento_tipo', 'padron_tipo', 'eleccion_tipo', 'cargo_id', 'lista_numero', 'lista_nombre','seccionprovincial_id','seccion_nombre','seccionprovincial_nombre','año','seccion_id','circuito_id','circuito_nombre','mesa_tipo'], axis=1)
```

```python
df = df.drop(df[df['cargo_nombre'] != 'PRESIDENTE Y VICE'].index) #elimino las filas donde el cargo que se vota no es presidente y vice
```

```python
df.head()
```

```python
df['distrito_nombre'].unique() #veo como están escritos lod nombres de cada provincia y región
```

```python
def codigo_provincia(valor):
        if valor == 'Ciudad Autónoma de Buenos Aires':
           return 'AR-C'
        elif valor == 'Buenos Aires':
            return 'AR-B'
        elif valor == 'Catamarca':
            return 'AR-K'
        elif valor == 'Neuquén':
            return 'AR-Q'
        elif valor == 'San Luis':
            return 'AR-D'
        elif valor == 'Córdoba':
            return 'AR-X'
        elif valor == 'La Pampa':
            return 'AR-L'
        elif valor == 'Mendoza':
            return 'AR-M'
        elif valor == 'Entre Ríos':
            return 'AR-E'
        elif valor == 'Corrientes':
            return 'AR-W'
        elif valor == 'Río Negro':
            return 'AR-R'
        elif valor == 'Salta':
            return 'AR-A'
        elif valor == 'Santiago del Estero':
            return 'AR-G'
        elif valor == 'Santa Fe':
            return 'AR-S'
        elif valor == 'Misiones':
            return 'AR-N'
        elif valor == 'La Rioja':
            return 'AR-F'
        elif valor == 'Chaco':
            return 'AR-H'
        elif valor == 'Formosa':
            return 'AR-P'
        elif valor == 'Chubut':
            return 'AR-U'
        elif valor == 'Jujuy':
            return 'AR-Y'
        elif valor == 'San Juan':
            return 'AR-J'
        elif valor == 'Tucumán':
            return 'AR-T'
        elif valor == 'Tierra del Fuego':
            return 'AR-V'
        elif valor == 'Santa Cruz':
            return 'AR-Z'
```

```python
df["distrito_id"] = df["distrito_nombre"].apply(lambda x: codigo_provincia(x)) #defino una función que genera una columna con el id correspondiente al mapa svg con el que voy a hacer el mapa coroplético
```

```python
df.head()
```

```python
df.drop(['mesa_id','agrupacion_id', 'cargo_nombre'], axis=1, inplace=True) 
```

```python
df.head()
```

```python
df.to_csv('DATASETS/Resultados_Generales_2023.csv', index=False)
```
